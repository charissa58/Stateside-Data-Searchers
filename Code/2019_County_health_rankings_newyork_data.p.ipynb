```python
import pandas as pd
import numpy as np
```

```python
# Reading in Data
df = pd.read_csv('2019_county_health_dataset.csv')
df
```

```python
# Selecting the columns we only want
select_df = df[['FIPS', 'County', 'Deaths', '% Fair or Poor Health', '% Low birthweight', 
            'Food Environment Index', '# Driving Deaths', '# Uninsured', 'Mental Health Provider Ratio']]
select_df
```

```python
# Removing the first row that was NaN 
select_df = select_df.dropna(how='all')
select_df
```

```python

select_df.head(60)
```

```python
# Removing the row that was NaN 
select_df = select_df.drop(labels=21, axis=0)
select_df
```

```python
# Removing the row that was NaN
# Also drop First Row (NY State)
select_df = select_df.drop(labels=62, axis=0)
select_df = select_df.drop(labels=0, axis=0)

select_df
```

```python
select_df.head(22)
```

```python
# The the year for the dataset
#select_df["Years"] = "2019"
select_df["Years"] = 2019

select_df
```

```python
# Reorganizing the dataset
clean_data = select_df[['Years', 'FIPS', 'County', 'Deaths', '% Fair or Poor Health', '% Low birthweight', 
            'Food Environment Index', '# Driving Deaths', '# Uninsured', 'Mental Health Provider Ratio']]
clean_data
```

```python
#Drop row 50 (NaN)

# Removing the row that was NaN
# Also drop First Row (NY State)
clean_data = clean_data.drop(labels=50, axis=0)
clean_data
```

```python
clean_data.dtypes
```

```python
# Spliting the Mental Health Provider Ratio 
changing = clean_data['Mental Health Provider Ratio'].str.split(':', expand=True)
changing.dtypes
```

```python
# Converting the string into int
changing = changing.astype('int')
```

```python
# Changing the Mental Health Provider Ratio into a int value
clean_data['Mental Health Provider Ratio'] = changing[1]/changing[0]
clean_data
```

```python
clean_data.shape
```

```python
clean_data.head(60)
```

```python
#convert all objects to integer
df = pd.DataFrame(clean_data)
df = df.astype({"FIPS":"int",
                        "Years":"object",
                        "Deaths":"int",
                        "% Fair or Poor Health":"int",
                        "% Low birthweight":"int",
                        "# Driving Deaths":"int",
                        "# Uninsured":"int"})
```
